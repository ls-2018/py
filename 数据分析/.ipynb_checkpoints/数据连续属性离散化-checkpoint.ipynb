```python
# 等宽法、等频法
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 等宽法    
ages = [20,22,25,27,21,23,37,31,61,45,41,32,80]
bins = [18,25,35,60,100]
cats = pd.cut(ages,bins,labels=list('ABCD'),right=True)# 默认为右闭
print(cats,type(cats))
print(cats.codes)# 区间等级
print(pd.value_counts(cats))# 按照区间计数


df = pd.DataFrame({'ages':ages})
# group_names = ['A','B','C','D']
s=pd.cut(df['ages'],bins)
df['ages_a']=s
cut_counts = s.value_counts(sort = False)

plt.scatter(s.index,df['ages'],cmap='Reds',c=cats.codes)
plt.grid()
```

```python
# 等频法,以相同数量的记录放入每个区间

data = np.random.rand(100)*10
s = pd.Series(data)
cats = pd.qcut(s,[0,0.1,0.5,0.9,1])  # 按照指定的规则划分;按照4分 
print(cats.head())
plt.scatter(s.index,s,cmap='Greens',c=pd.qcut(data,4).codes)# 点的颜色有4中
plt.xlim([0,100])
plt.grid()
```
