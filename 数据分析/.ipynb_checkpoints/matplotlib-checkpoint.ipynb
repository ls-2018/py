```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

labels = ['man', 'female', 'unknow']

X = [20, 23, 23]
fig = plt.figure()
plt.pie(X, labels=labels, autopct='%1.2f%%')# 圆

plt.title('pie chart')
#plt.show()
plt.savefig('PieChart.jpg')
```
