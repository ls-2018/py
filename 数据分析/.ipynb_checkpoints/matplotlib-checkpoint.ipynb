```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

labels = ['man', 'female', 'unknow']

X = [20, 23, 23]
fig = plt.figure()
plt.pie(X, labels=labels, autopct='%1.2f%%')# 圆

plt.title('pie chart')
#plt.show()
plt.savefig('PieChart.jpg')
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

x= np.arange(-10,10,5)
plt.plot(x,np.arange(-10,10,5),'o-',color='red',label='y=x')
 
y2=x**2
plt.plot(x,y2,'^:',color='blue',label='y=x*x')

y3=3*x**3 +5*x**2 +2*x+1
plt.plot(x,y3,'*:',color='black',label='y=3*x**3 +5*x**2 +2*x+1')


plt.legend()
plt.show()













```
