```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

labels = ['man', 'female', 'unknow']

X = [20, 23, 23]
fig = plt.figure()
plt.pie(X, labels=labels, autopct='%1.2f%%')# 圆

plt.title('pie chart')
#plt.show()
plt.savefig('PieChart.jpg')
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

x= np.arange(-10,10,5)
plt.plot(x,np.arange(-10,10,5),'o-',color='red',label='y=x')
 
y2=x**2
plt.plot(x,y2,'^:',color='blue',label='y=x*x')

y3=3*x**3 +5*x**2 +2*x+1
plt.plot(x,y3,'*:',color='black',label='y=3*x**3 +5*x**2 +2*x+1')


plt.legend()
plt.show()













```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
 
fig = plt.figure() # 画布

# 调节子图间距
# plt.subplots_adjust(left=2,bottom=2,right=20,top=20,wspace=1,hspace=1)

ax1 = fig.add_subplot(2,2,1)
ax1.pie([20, 23, 23], labels=['man', 'female', 'unknow'], autopct='%1.2f%%')

ax2 = fig.add_subplot(2,2,2)
ax2.pie([20, 23, 23], labels=['man', 'female', 'unknow'], autopct='%1.2f%%')


plt.title('pie chart')

fig.show()
#fig.savefig('PieChart.jpg')

```
