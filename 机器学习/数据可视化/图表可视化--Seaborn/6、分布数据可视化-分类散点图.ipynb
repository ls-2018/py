```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# stripplot 按照不同类别对样本数据进行分布散点图绘制
tips = sns.load_dataset('tips')
print(tips.head())

sns.stripplot(x='day',# 设置分组统计字段
             y='total_bill',
              data = tips,
              jitter=True,# 当数据重合较多时，会自动调整   0-1之间
              size=5,edgecolor='w',
              linewidth=1,
              marker='o', 
             )
```

```python
# stripplot 按照不同类别对样本数据进行分布散点图绘制
tips = sns.load_dataset('tips')
print(tips.head())

sns.stripplot(x='sex',# 设置分组统计字段
             y='total_bill',
              data = tips,
              jitter=True,# 当数据重合较多时，会自动调整   0-1之间
#               linewidth=1,
#               marker='o',   
              hue='day',# 做了二次分组

             )
```

```python
tips = sns.load_dataset('tips')
print(tips.head())

sns.stripplot(x='sex',# 设置分组统计字段
             y='total_bill',
              data = tips,
              jitter=True,# 当数据重合较多时，会自动调整   0-1之间
#               linewidth=1,
#               marker='o',   
              hue='day',# 做了二次分组
                dodge=True,# 是否拆分
             )
```

```python
tips = sns.load_dataset('tips')
print(tips.head())

sns.stripplot(x='day',# 设置分组统计字段
             y='total_bill',
              data = tips,
              jitter=True,# 当数据重合较多时，会自动调整   0-1之间 
              order=['Sat','Sun'],
             )
```

```python
#swarmplot  分出散点图
sns.swarmplot(x='total_bill',
             y='day',
             data=tips,
             size=5,
             edgecolor='w',
             palette='Reds'
             )
```
