```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#  线性回归
tips = sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',
           data = tips,palette='Set1',
           ci=70,# 误差值
           size = 5,# 图标大小
            hue='smoker',
            markers=['+','o'],# 点样式
          )
plt.grid()
```

```python
# col  拆分成多个表格
sns.lmplot(x='total_bill',y='tip',col='smoker',data=tips)   # 是否吸烟 与  小费总额的关系
```

```python
# col  拆分成多个表格
sns.lmplot(x='size',y='total_bill',
           hue='smoker',
#            col='day',
           data=tips,
           aspect=0.6,# 长宽比
           x_jitter=30,# 给x或者y轴随机增加噪点
#            col_wrap=4,# 每行的列数
          )   # 是否吸烟 与  小费总额的关系
```

```python
# col  拆分成多个表格
sns.lmplot(x='size',y='total_bill',
           hue='smoker',
           col='day',
           data=tips,
           aspect=0.6,# 长宽比
           x_jitter=30,# 给x或者y轴随机增加噪点
           col_wrap=4,# 每行的列数
          )   # 是否吸烟 与  小费总额的关系
```

```python
#  非线性回归
sns.lmplot(x='total_bill',y='tip',
          data=tips,
          order=2 , #  最高阶次为2 的回归方程
          )

```
