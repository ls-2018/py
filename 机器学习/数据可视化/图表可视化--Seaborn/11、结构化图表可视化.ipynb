```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# 绘制直方图
tips = sns.load_dataset('tips')
g = sns.FacetGrid(tips,col='time',row='smoker')    #  列、行

g.map(plt.hist,'total_bill',alpha=0.5,color='k',bins=10)
```

```python
# 分割
tips = sns.load_dataset('tips')
g = sns.FacetGrid(tips,col='time',hue='smoker' )    #  列、行

g.map(plt.scatter,
      'total_bill',
      'tip',
      edgecolor='w',
      s=40,
      linewidth=1
     )
g.add_legend()
```

```python
# 图表矩阵



attend = sns.load_dataset('attention')
print(attend.head())
print(attend.shape)
g = sns.FacetGrid(attend,col='subject',col_wrap=5,# 每行的图表数量
                  size=1.5
                 )
g.map(plt.plot,'solutions','score',
     marker='o',
     color='gray',
      linewidth=2
     )
# 绘制矩阵图表

g.set(xlim=(0,4),
     ylim=(0,10),
      xticks=[0,1,2,3,4],
      yticks=[0,2,4,6,8,10]
     )
# 设置X，y轴热度
```
