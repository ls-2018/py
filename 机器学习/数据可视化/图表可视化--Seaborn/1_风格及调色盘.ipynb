```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
# 创建正弦函数以及图表

def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
sinplot()
# 创建正弦函数，以及出图
```

```python
sns.set()   # 一旦设置，全局都回更改，只有重启可以修改
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
sinplot()
plt.grid(linestyle='--')
# 创建正弦函数，以及出图
```
