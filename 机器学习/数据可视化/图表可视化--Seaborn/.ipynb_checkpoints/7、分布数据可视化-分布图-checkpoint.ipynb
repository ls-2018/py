```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# 1、boxplot（）箱型图

tips = sns.load_dataset('tips')
print(tips.head())
sns.boxplot(x='day',y='total_bill',
           data=tips,
           linewidth=2,
           fliersize=3,# 异常点大小
           palette='hls',# 调色板
           whis=1.5,# 设置IQR
           notch =True,# 是否以中值做凹陷
           order = ['Thur','Fri','Sat','Sun'],# 帅选类别 
           )

sns.swarmplot(x='day',y='total_bill',data=tips,color='k',size=3,alpha=0.3)
```

```python
# 1、boxplot（）箱型图
tips = sns.load_dataset('tips')
print(tips.head())
sns.boxplot(x='day',y='total_bill',
           data=tips,
           hue='smoker',
           palette='Reds',
           )

sns.swarmplot(x='day',y='total_bill',data=tips,color='k',size=3,alpha=0.3)
```

```python
# 2 小提琴图  violinplot    
"""
linewidth
width   箱之间的间隔比例
palette  调色板
order
"""
sns.violinplot(
    x='day',y='total_bill',data=tips,
    scale='area',# 小提琴的宽度，area->面积相同，count---->按照样本数量决定宽度，width-->宽度一样
    gridsize=50,# 小提琴图变现的平滑度，越高越平滑
    inner='box',# 设置内部显示类型   box       quartile  point    stick  None
    abx=0.8 ,# 控制拟合程度，一般不可以设置
)
```
