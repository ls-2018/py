```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set_context('notebook')
```

```python
# 直方图
rs = np.random.RandomState(10)# 设定随机数种子
s = pd.Series(rs.randn(100)*100)

"""
s
bins箱数
hist，ked     是否显示箱、密度曲线
norm_hist     直方图是否按照密度来显示
rug   是否显示数据分布情况
vertical   是否水平显示
axlabel   
"""
sns.distplot(s,bins =20,hist=True,kde=True,norm_hist=False,
            rug=True,vertical=False,
            color='g',label='displot',
            axlabel='x')
plt.legend()
```

```python
sns.distplot(s,rug=True,
             rug_kws={'color':'g'},
             kde_kws={'color':'k','lw':1,'label':'KDE','linestyle':'--'},
             hist_kws={'histtype':'step','linewidth':1,'alpha':1,'color':'g'}
)
```

```python
# 密度图 kdeplot
# 一维

sns.kdeplot(s,shade=False,# 是否填充
            color = 'r',
            vertical=False,   #是否水平
           )
sns.kdeplot(s,bw=5,label='bw:0.2',
           linestyle='-',
            linewidth=1.2,
            alpha=0.5
           )
sns.kdeplot(s,bw=20,label='bw:2',
           linestyle='-',
            linewidth=1.2,
            alpha=0.5
           )
# bw   控制拟合的程度，类似直方图的箱数

sns.rugplot(s,height=0.1,color='k',alpha=0.5)# 频率分布图      透明度
```

```python
# 二维
rs = np.random.RandomState(2)# 设定随机数种子
df = pd.DataFrame(rs.randn(100,2)*100,
                  columns=['A','B'])

sns.kdeplot(df['A'],df['B'],
            cbar=True,  # 是否显示颜色图例,右侧
            shade=True,# 是否填充
            cmap='Reds',
            shade_lowest=False,  # 外围颜色是否显示
            n_levels=200,# 曲线个数
           )  
# bw   控制拟合的程度，类似直方图的箱数
sns.rugplot(df['A'],height=0.1,color='g',axis='x',alpha=0.5)# 频率分布图      透明度
sns.rugplot(df['B'],height=0.1,color='r',axis='y',alpha=0.5)# 频率分布图      透明度
```
