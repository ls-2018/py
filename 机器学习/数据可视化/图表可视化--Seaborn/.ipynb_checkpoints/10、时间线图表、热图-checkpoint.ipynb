```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# 1 时间线图表   tsplot()
x = np.linspace(0,15,31)
data = np.sin(x) + np.random.rand(10,31) + np.random.rand(10,1)

print(data.shape)
print(pd.DataFrame(data).head())

# 创建数据
sns.tsplot(data=data,
          err_style='ci_band',#   误差数据风格，  ci_band  ci_bars   boot_traces  boot_kde  unit_traces  unitpoints
          interpolate=True,#  是否连线
           ci=[40,70,90],# 设置误差区间
           color='g',# 设置颜色
           n_boot=300,# 迭代次数，--线数
          )
```

```python
#  参数设置

gammas = sns.load_dataset('gammas')
print(gammas.head())

sns.tsplot(time='timepoint',# 时间数据 x轴
           value='BOLD signal',# y轴value
           unit='subject',# 
           condition='ROI',# 分类
           data= gammas
          )
```

```python
# 2 热图
df = pd.DataFrame(np.random.rand(10,12))
print(df.shape)
sns.heatmap(df,
            vmin=0,
            vmax=1,#  设置图例最大最小值
           )
```

```python
# 参数设置
flights = sns.load_dataset('flights')
flights = flights.pivot('month','year','passengers')

sns.heatmap(
    flights,
    annot=True,# 是否显示数值
    fmt='d',# 格式化字符串
    linewidths=0.2,#各自边线宽度
#     center=100,# 调色盘的色彩中心值，若没有指定，则以cmap为主
    cmap='Reds',
    cbar=True,# 显示图例色带
#     cbar_kws={'orientation':'horizontal'}#  横向显示图例色带
#     square=True,# 是否正方形显示图表
)
```

```python
# 绘制单边热图
rs = np.random.RandomState(33)
df = pd.DataFrame(np.random.rand(100,20))
corr = df.corr()# 求解相关性矩阵表格

mask = np.zeros_like(corr,dtype=np.bool)
mask[np.tril_indices_from(mask)] =True
# 设置一个上三角形  蒙板

cmap = sns.diverging_palette(200,10,as_cmap=True)
# 设置调色盘

sns.heatmap(corr,mask=mask,cmap=cmap,vmax=3,center=0,
           square=True,linewidths=0.2)
# 生成半边热图
```
