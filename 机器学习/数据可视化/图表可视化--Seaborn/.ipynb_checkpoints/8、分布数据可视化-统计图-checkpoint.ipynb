```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# 1 、 barplot()
# 柱状图 - 置信区间估计
# 置信区间  样本均值+ 抽样误差

titanic = sns.load_dataset('titanic')
# print(titanic.head())
# 加载数据

sns.barplot(x='sex',y='survived',hue='class',data=titanic,
           palette='hls',
            order=['male','female'],# 帅选类型
            capsize=0.05,# 误差线横向延伸宽度
            saturation=.8,# 颜色饱和度
            errcolor='k',
            errwidth=2,   # 误差线颜色，宽度
            ci='sd',# 置信区间误差   0-100内值， sd   None  
           )
print(titanic.groupby(['sex','class']).mean()['survived'])
print(titanic.groupby(['sex','class']).std()['survived'])
```

```python

tips = sns.load_dataset('tips')

sns.barplot(x='day',y='total_bill',hue='sex',data=tips,
           palette='Blues',
            capsize=0.05,# 误差线横向延伸宽度
            saturation=.8,# 颜色饱和度
            edgecolor='w',
           )
print(tips.groupby(['day','sex']).mean())
```

```python
# barplot  柱状图 --- 置信区间估计

crashes = sns.load_dataset('car_crashes').sort_values('total',ascending=False)

f,ax = plt.subplots(figsize=(6,15))

sns.set_color_codes('pastel')
sns.barplot(x='total',y='abbrev',data=crashes,label='Toal',color='b',edgecolor='w')
# 设置一个柱状图

sns.set_color_codes('muted')
sns.barplot(x='alcohol',y='abbrev',data=crashes,
           label='Alcohol-involved',color='b',edgecolor='w')
# 设置第二个柱状图
ax.legend(ncol=2,loc='lower right')
sns.despine(left=True,bottom=True)

```

```python
# 计数柱状图 
# sns.countplot(x='class',hue='who',data=titanic,palette='magma')
sns.countplot(y='class',hue='who',data=titanic,palette='magma')
```

```python
# 折线图  pointplot
sns.pointplot(x='time',y='total_bill',hue='smoker',data=tips,
             palette='hls',
             dodge=True,# 设置点是否分开
             join=False,# 是否连线
              markers=['o','x'],linestyles=['-','--'],# 设置点样式，线型
             )
tips.groupby(['time','smoker']).mean()['total_bill']
```
