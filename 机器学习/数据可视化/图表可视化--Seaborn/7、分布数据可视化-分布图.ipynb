```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# 1、boxplot（）箱型图

tips = sns.load_dataset('tips')
print(tips.head())
sns.boxplot(x='day',y='total_bill',
           data=tips,
           linewidth=2,
           fliersize=3,# 异常点大小
           palette='hls',# 调色板
           whis=1.5,# 设置IQR
           notch =True,# 是否以中值做凹陷
           order = ['Thur','Fri','Sat','Sun']# 帅选类别 
           )

sns.swarmplot(x='day',y='total_bill',data=tips,color='k',size=3,alpha=0.3)
```
